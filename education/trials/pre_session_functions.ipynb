# Funciones en Python

Las **funciones** permiten encapsular lógica y reutilizar código. Se definen con la palabra clave `def` seguida del nombre y la lista de argumentos. El resultado se devuelve con `return`.

```python
def nombre_funcion(param1, param2):
    # bloque de código
    resultado = param1 + param2
    return resultado
```

Definir funciones es esencial para construir pipelines modulares y reutilizables.


````python
# Ejemplo de función que calcula el promedio de una lista de números
def calcular_promedio(numeros):
    suma = sum(numeros)
    cantidad = len(numeros)
    promedio = suma / cantidad
    return promedio

valores = [2, 4, 6, 8]
print('El promedio es:', calcular_promedio(valores))
````

## Parámetros y valores de retorno

Puedes proporcionar valores por defecto a los parámetros (`param=valor_por_defecto`), y una función puede devolver múltiples valores utilizando tuplas.


````python
# Ejemplo de función con valor por defecto y múltiples retornos
def calcular_estadisticas(lista, mostrar=False):
    media = sum(lista) / len(lista)
    minimo = min(lista)
    maximo = max(lista)
    if mostrar:
        print('Media:', media, 'Mínimo:', minimo, 'Máximo:', maximo)
    return media, minimo, maximo

nums = [5, 10, 3, 8]
resultado = calcular_estadisticas(nums, mostrar=True)
print('Tupla devuelta:', resultado)
````
