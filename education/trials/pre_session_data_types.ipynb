# Variables y Tipos de Datos

Las **variables** almacenan valores que tu programa puede utilizar. En Python no es necesario declarar el tipo antes de asignar un valor. Algunos tipos básicos son:

- **Enteros (`int`)**: números sin parte decimal (por ejemplo `5`, `-3`).
- **Flotantes (`float`)**: números con parte decimal (por ejemplo `3.14`, `-0.5`).
- **Cadenas (`str`)**: texto entre comillas (por ejemplo `"hola"`).
- **Booleanos (`bool`)**: valores lógicos `True` o `False`.

Puedes averiguar el tipo de una variable con la función `type()`.


````python
# Ejemplos de variables de distintos tipos
edad = 30             # entero
altura = 1.70         # flotante
nombre = 'Ana'        # cadena
es_mayor = edad > 18  # booleano (resultado de una comparación)

print('Edad:', edad, '->', type(edad))
print('Altura:', altura, '->', type(altura))
print('Nombre:', nombre, '->', type(nombre))
print('¿Es mayor de edad?', es_mayor, '->', type(es_mayor))
````

## Conversión de tipos

A veces necesitas convertir un valor de un tipo a otro. Por ejemplo, transformar una cadena `"42"` en el número `42` usando `int("42")` o convertir un número en cadena usando `str(42)`.


````python
# Ejemplo de conversión
peso_texto = '72.5'
# Convertir a flotante
peso = float(peso_texto)
print('Peso como número:', peso)
print('Tipo de dato:', type(peso))
# Volver a cadena
peso_str = str(peso)
print('Peso como cadena:', peso_str)
print('Tipo de dato:', type(peso_str))
````
