## Diferencia entre script y notebook

Un *script* es un archivo de texto con instrucciones que se ejecutan de principio a fin, mientras que un *notebook* está dividido en celdas que pueden ejecutarse de manera independiente. Esto permite explorar los datos de forma iterativa, documentar los pasos y conservar resultados intermedios.
