# Control de Flujo: Condicionales y Bucles

Las **estructuras de control** permiten tomar decisiones y repetir acciones. En Python, las principales estructuras de flujo son las condicionales (`if`, `elif`, `else`) y los bucles (`for`, `while`).

## Condicionales

Una sentencia `if` evalúa una expresión booleana y ejecuta un bloque de código sólo si la condición se cumple. Puedes usar `elif` y `else` para evaluar condiciones adicionales o un caso por defecto.

## Bucles

- **`for`**: itera sobre los elementos de una secuencia (lista, cadena, etc.).
- **`while`**: repite un bloque de código mientras se cumpla una condición.

Los bucles permiten procesar cada elemento de una lista o repetir una tarea hasta alcanzar un criterio.


````python
# Ejemplo de condicional
x = 10
if x > 5:
    print('x es mayor que 5')
elif x == 5:
    print('x es igual a 5')
else:
    print('x es menor que 5')

# Ejemplo de bucle for
edades = [45, 60, 30, 55]
conteo = 0
for edad in edades:
    if edad > 50:
        conteo += 1
print('Número de pacientes mayores de 50 años:', conteo)

# Ejemplo de bucle while
contador = 0
while contador < 3:
    print('Iteración', contador)
    contador += 1
````
